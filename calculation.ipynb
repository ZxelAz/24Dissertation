```python
import numpy as np
from scipy.integrate import quad

s = 0.21

def f(x, d):
    return (np.exp(-(x - d / 2)**2 /(2*s**2)) + np.exp(-(x + d / 2)**2) /(2*s**2) )/(2*s * (2*np.pi)**0.5)

def df(x, d):
    return ((x - d / 2) * np.exp(-(x - d / 2)**2 /(2*s**2)) - (x + d / 2) * np.exp(-(x + d / 2)**2 /(2*s**2)))/((2*np.pi)**0.5 * 2*s**3)

def integrand(x, d):
    return df(x, d)**2 / f(x, d)

def integrate_integrand(d):
    # Use a large finite range instead of -inf to +inf
    large_range = 10
    result, error = quad(integrand, -15, +15, args=(d,))
    return result, error

# Example usage:
d = 0.3
result, error = integrate_integrand(d)
print(f"Integration result: {result}, with error estimate: {error}")
```

```python
import matplotlib.pyplot as plt
d_values = np.linspace(0, 8, 100)
integral_values = [integrate_integrand(d)[0] for d in d_values]

# Plotting the results
plt.figure(figsize=(10, 6))
plt.plot(d_values, integral_values)
plt.xlabel('d')
plt.ylabel('FI')
plt.title('Fisher Information')
plt.legend()
plt.grid(True)
plt.show()
```
